```python
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits import mplot3d
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt      
import seaborn as sns
```

```python
df = pd.read_csv('laptops3.csv', header=0, sep=',') 
df2 = pd.read_csv('laptops2.csv', header=0, sep=',') 
del df['Unnamed: 0']
df
```

```python
df.Price.hist(bins=50)
plt.xlabel('Price', color ='teal')
plt.ylabel('laptops', color ='teal')
```

```python
sns.violinplot(x='RAM', y='Storage', data=df)
plt.title('Violin plot of x and y')
plt.show()
```

```python
fig = plt.figure(figsize=(10,3))
sns.boxplot(df.Storage)
```

```python
plt.scatter(x='RAM', y='Storage', data=df, c='Price', cmap='coolwarm')
plt.colorbar()

# Add axis labels and a title
plt.xlabel('RAM (GB)')
plt.ylabel('Storage (GB)')
plt.title('RAM and Storage vs. Price')
```

```python
fig = plt.figure(figsize=(10,3))
sns.boxplot(df.Price)
```

```python
df2 = pd.DataFrame(df, columns = ['Graphic Card','CPU', 'RAM', 'Storage', 'Price'])
sns.heatmap(df2.corr(), annot=True, cmap='Greens')
```

```python
sns.countplot(x='RAM', data=df)
plt.title('Count of values in column RAM')
plt.figure(figsize=(30,5))
plt.show()
```

```python
df['Storage'] = df['Storage'].astype(int)
sns.countplot(x='Storage', data=df)
plt.title('Count of values in column Storage')
plt.show()
```

```python
df['Touch Screen'].value_counts().plot(kind = 'pie')
```

```python
sns.scatterplot(x='CPU', y='Price', data=df)
plt.title('Scatter plot of x, y')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
xdata = df.Storage
ydata = df.RAM
zdata = df.Price
plt.xlabel('Storage')
plt.ylabel('RAM')
ax.scatter3D(xdata,ydata,zdata,c=zdata,depthshade= False )
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['RAM'], df['Storage'], df['Gen'], c=df['Price'], cmap='Reds')
ax.set_xlabel('RAM')
ax.set_ylabel('Storage')
ax.set_zlabel('Gen')

plt.show()
```

```python
cpu_price_pivot = df2.pivot_table(index='CPU', values='Price', aggfunc='mean')
sns.heatmap(cpu_price_pivot, annot=True, cmap='coolwarm', fmt='.2f', yticklabels=cpu_price_pivot.index)
```
